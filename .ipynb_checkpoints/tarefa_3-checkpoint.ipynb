```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

# Euler Implícito

$x_{k+1} = x_{k} + \Delta f(t_{k+1}, x_{k+1})$
k = 0,1,2,...,n-1:

\begin{array}{lcl} 
t_{k+1} & = & t_{k} + \Delta t \\ 
X_{k+1} & \leftarrow & RESOLVEX[X - X_{k} - \Delta t f(t_{k+1}, X) = 0]
\end{array}

# Método do Trapézio

$x_{k+1} = x_{k} + \Delta f(t_{k+1}, x_{k+1})$
k = 0,1,2,...,n-1:

\begin{array}{lcl} 
t_{k+1} & = & t_{k} + \Delta t \\ 
X_{k+1} & \leftarrow & RESOLVEX[X - X_{k} - \dfrac{\Delta t}{2} ( f(t_{k}, X_{k}) + f(t_{k+1}, X) ) = 0]
\end{array}

# Newton

$$
J(X_{k+1}^{[j]})\Delta X^{[j+1]} = - g(X_{k+1}^{[j]})
$$

A melhor aproximação para $X_{k+1}^{[0]}$ é $X_{k}$

```python
##===========================================================
## Método de Newton Bidimensional
##===========================================================
def newton2D(x, y, xk, yk, delta, a1, b1, a2, b2, Jg, g, IN):
    for i in range(IN):
        deltaX = np.linalg.solve(Jg(x, y, a1, b1, a2, b2, delta), -g(x, y, xk, yk, a1, b1, a2, b2, delta))
        x = x + deltaX[0]
        y = y + deltaX[1]
    
    return np.array([x, y])
```

Vamos implementar uma função genérica para métodos de passo único bidimensionais. Vale lembrar que a melhor aproximação para o $X_{k+1}^{[0]}$ que é argumento da função $RESOLVERX$ - que, neste trabalho, é o método de newton bidimensional - é o próprio $X_{k}$.

Note também que a diferença entre o Método de Euler Implícito e o Método do Trapézio está na definição da função $g$, da qual devemos encontrar os zeros para estimar o valor da função em determinado passo.

```python
##===========================================================
## FUNÇÃO GENÉRICA PARA RESOLUÇÃO DE ODE'S POR PASSO ÚNICO
## Assinatura do método:
## t0: tempo inicial
## tf: tempo final
## n: número de passos de integração, usado para calcular delta = (tf - t0)/n
## X0: vetor com as condições iniciais
## RESOLVERX: função para resolução de equações algébricas bidimensionais
## a1, b1, a2, b2: parâmetros da equação
## Jg: função que retorna a Jacobiana de g
## g: função que representa a equação algébrica a ser resolvida em RESOLVERX
## IN: número de iterações do método RESOLVERX
##===========================================================
def ode_passo_unico_bidimensional(t0, tf, n, X0, RESOLVEX, a1, b1, a2, b2, Jg, g, IN):
    h = (tf - t0)/n
    t = t0
    T = [t]
    X_t = [X0]
    for k in range(n):
        t = t + h
        T.append(t)
        Xk = X_t[-1]        
        X_next_inicial = Xk
        X_next = RESOLVEX(X_next_inicial[0], X_next_inicial[1], Xk[0], Xk[1], h, a1, b1, a2, b2, Jg, g, IN)
        X_t.append(X_next)
    return T, X_t
```

# O Problema Manufaturado

### Solução Manufaturada

Exemplo Bidimensional. Sistema:

\begin{array}{lcl}
    \dfrac{dx_{1}}{dt} = \dfrac{-8}{5}x_{1}(t) + \dfrac{3}{10}x_{2}(t) \\
    \dfrac{dx_{2}}{dt} = \dfrac{8}{5}x_{1}(t) - \dfrac{4}{5}x_{2}(t) \\
    x_{1}(0) = 0 \\
    x_{2}(0) = 12
\end{array}

Solução

$$
\left[
\begin{matrix}
     x(t) \\
     y(t)
\end{matrix}
\right]
= 
\left[
\begin{matrix}
     \dfrac{9}{4}e^{(-2/5)t} - \dfrac{9}{4}e^{-2t} \\
     9e^{(-2/5)t} + 3e^{-2t}
\end{matrix}
\right]
$$

### Manufaturada

Abaixo, definimos a solução analítica:

$X(t) = \left[
\begin{matrix}
     x(t) \\
     y(t)
\end{matrix}
\right]$

```python
def manufaturada(t):
    linha1 = 9/4.0 * math.exp(-2/5.0 * t) - 9/4.0 * math.exp(-2*t)
    linha2 = 9*math.exp(-2/5.0 * t) + 3*math.exp(-2*t)
    return np.array([linha1, linha2])
```

Vamos definir os parâmetros do problema manufaturado

```python
a1_manufaturada, b1_manufaturada, a2_manufaturada, b2_manufaturada = -8/5.0, 3/10.0, 8/5.0, -4/5.0 
x0_manufaturada, y0_manufaturada = 0, 12
```

Agora, vamos definir as funções utilizadas nos cálculos dos zeros via método de Newton Bidimensional

```python
#==================================================================
# Funções das equações algébricas do problema manufaturado com Euler Implícito:
#
# g_manufaturada é a equação algébrica para a qual queremos encontrar as raízes
# g1_manufaturada e g2_manufaturada são componentes de g_manufaturada
# Jg_manufaturada é a função que retorna a matriz jacobiana de g_manufaturada
#==================================================================

def g1_manufaturada(x, y, xk, alpha1, beta1, delta):
    return x - xk - delta*(alpha1*x + beta1*y)

def g2_manufaturada(x, y, yk, alpha2, beta2, delta):
    return y - yk - delta*(alpha2*x + beta2*y)

def g_manufaturada(x, y, xk, yk, alpha1, beta1, alpha2, beta2, delta):
    return np.array([g1_manufaturada(x, y, xk, alpha1, beta1, delta), 
                     g2_manufaturada(x, y, yk, alpha2, beta2, delta)])

#Possível Evolução: Jg específica -> Jg genérica
def Jg_manufaturada(x, y, alpha1, beta1, alpha2, beta2, delta):
    linha1 = [1 - delta*alpha1, -delta*beta1]
    linha2 = [-delta*alpha2, 1 - delta*beta2]
    return np.array([linha1, linha2])
```

```python
#==================================================================
# Funções das equações algébricas do problema manufaturado com Método do Trapézio:
#
# g_manuf_trapezio é a equação algébrica para a qual queremos encontrar as raízes
# g1_manuf_trapezio e g2_manuf_trapezio são componentes de g_manuf_trapezio
# Jg_manuf_trapezio é a função que retorna a matriz jacobiana de g_manuf_trapezio
#==================================================================

def g1_manuf_trapezio(x, y, xk, yk, alpha1, beta1, delta):
    return x - xk - (delta/2.0)*((alpha1*xk + beta1*yk) + (alpha1*x + beta1*y))

def g2_manuf_trapezio(x, y, xk, yk, alpha2, beta2, delta):
    return y - yk - (delta/2.0)*((alpha2*xk + beta2*yk) + (alpha2*x + beta2*y))

def g_manuf_trapezio(x, y, xk, yk, alpha1, beta1, alpha2, beta2, delta):
    return np.array([g1_manuf_trapezio(x, y, xk, yk, alpha1, beta1, delta), 
                     g2_manuf_trapezio(x, y, xk, yk, alpha2, beta2, delta)])

#Possível Evolução: Jg específica -> Jg genérica
def Jg_manuf_trapezio(x, y, alpha1, beta1, alpha2, beta2, delta):
    linha1 = [1 - (delta/2.0)*alpha1, -(delta/2.0)*beta1]
    linha2 = [-(delta/2.0)*alpha2, 1 - (delta/2.0)*beta2]
    return np.array([linha1, linha2])
```

Vamos definir o intervalo de integração para o problema manufaturado como $[0, 10]$

```python
t0 = 0.0
tf = 10.0
```

```python
manufaturada(10)
```

## Método de Euler Implícito, problema manufaturado

```python
ns = [4, 8, 16, 32, 64, 128, 256, 512, 1024]
iteracoes = 100

Ts = []
Xs = []
erro_x = []
erro_y = []

analitica = manufaturada(tf)

for n in ns:
    T, X_ = ode_passo_unico_bidimensional(t0, tf, n, np.array([x0_manufaturada, y0_manufaturada]), 
                               newton2D, a1_manufaturada, b1_manufaturada, a2_manufaturada, 
                               b2_manufaturada, Jg_manufaturada, g_manufaturada, iteracoes)
    
    X_ = np.array(X_).T
    
    ex = math.fabs(X_[0][-1] - analitica[0])
    ey = math.fabs(X_[1][-1] - analitica[1])
    erro_x.append(ex)
    erro_y.append(ey)
    
    Ts.append(T)
    Xs.append(X_)
```

```python
plt.figure()

for T, X, delta in zip(Ts, Xs, [(tf - t0)/n for n in ns]):    
    plt.plot(T, X[0], label = '$\Delta t$ = '+str('%.3f'%(delta)))

plt.title('Euler Implícito: $x_{1}(t)$ Manufaturado')
plt.xlabel('t')
plt.ylabel('$x_{1}(t)$')

plt.legend()
plt.grid(True)
plt.show()
```

# Modelo Presa-Predador

\begin{array}{lcl} 
\dot{x}(t) & = & 0.84x(t) - 0.27x(t)y(t) \\ 
\dot{y}(t) & = & -0.38y(t) + 0.25y(t)x(t)
\end{array}

Dados $x(0) = 3.5$ e $y(0) = 2.7$

Vamos integrar a função em $[0,0]$.

$t_{0} = 0$ e $t_{f} = 40$

```python
t0 = 0
tf = 40
a1, b1, a2, b2 = 0.84, 0.27, 0.38, 0.25 
x0, y0 = 3.5, 2.7
```

```python
#============================================================================
# Funções das equações algébricas para o Método de Euler Implícito
# g_implicito é a equação algébrica para a qual queremos encontrar as raízes
# g1_implicito e g2_implicito são componentes de g_implicito
# Jg_implicito é a função que retorna a matriz jacobiana de g_implicito
#=============================================================================
def g1_implicito(x, y, xk, alpha1, beta1, delta):
    return x - xk - delta*(alpha1*x - beta1*x*y)

def g2_implicito(x, y, yk, alpha2, beta2, delta):
    return y - yk - delta*(-alpha2*y + beta2*x*y)

def g_implicito(x, y, xk, yk, alpha1, beta1, alpha2, beta2, delta):
    return np.array([g1_implicito(x, y, xk, alpha1, beta1, delta), 
                     g2_implicito(x, y, yk, alpha2, beta2, delta)])

def Jg_implicito(x, y, alpha1, beta1, alpha2, beta2, delta):
    linha1 = [1 - delta*(alpha1 - beta1*y), delta*beta1*x]
    linha2 = [-delta*beta2*y, 1 - delta*(-alpha2 + beta2*x)]
    return np.array([linha1, linha2])
```

```python
#======================================================================
# Funções das equações algébricas para o Método do Trapézio
# g_trapezio é a equação algébrica para a qual queremos encontrar as raízes
# g1_trapezio e g2_trapezio são componentes de g_trapezio
# Jg_trapezio é a função que retorna a matriz jacobiana de g_trapezio
#======================================================================
def g1_trapezio(x, y, xk, yk, alpha1, beta1, delta):
    return x - xk - (delta/2.0)*((alpha1*xk - beta1*xk*yk) + (alpha1*x - beta1*x*y))

def g2_trapezio(x, y, xk, yk, alpha2, beta2, delta):
    return y - yk - (delta/2.0)*((-alpha2*yk + beta2*xk*yk) + (-alpha2*y + beta2*x*y))

def g_trapezio(x, y, xk, yk, alpha1, beta1, alpha2, beta2, delta):
    return np.array([g1_trapezio(x, y, xk, yk, alpha1, beta1, delta), 
                     g2_trapezio(x, y, xk, yk, alpha2, beta2, delta)])

def Jg_trapezio(x, y, alpha1, beta1, alpha2, beta2, delta):
    linha1 = [1 - (delta/2.0)*(alpha1 - beta1*y), (delta/2.0)*beta1*x]
    linha2 = [-(delta/2.0)*beta2*y, 1 - (delta/2.0)*(-alpha2 + beta2*x)]
    return np.array([linha1, linha2])
```

### Presa-Predador com Método de Euler Implícito

Como já verificamos a corretude do código implementado via solução manufaturada, vamos utilizar um número de passos menor para a aproximação ao zero de $g$ via método de newton bidimensional

```python
ns = [4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768]
iteracoes = 5

Ts = []
Xs = []

for n in ns:
    T, X_ = ode_passo_unico_bidimensional(t0, tf, n, np.array([x0, y0]), 
                                          newton2D, a1, b1, a2, b2, 
                                          Jg_implicito, g_implicito, iteracoes)
    
    X_ = np.array(X_).T
    
    Ts.append(T)
    Xs.append(X_)
```

```python
plt.figure()

for T, X, delta in zip(Ts, Xs, [(tf - t0)/n for n in ns]):    
    plt.plot(T, X[0], label = '$\Delta t$ = '+str('%.3f'%(delta)))

plt.title('Euler Implícito: $x_{1}(t)$ Presa-Predador')
plt.xlabel('t')
plt.ylabel('$x_{1}(t)$')

plt.legend(bbox_to_anchor=(1,1))
plt.grid(True)
plt.show()
```

### Presa-Predador com Método do Trapézio

```python
ns = [4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768]
iteracoes = 5

Ts = []
Xs = []

for n in ns:
    T, X_ = ode_passo_unico_bidimensional(t0, tf, n, np.array([x0, y0]), 
                                          newton2D, a1, b1, a2, b2, 
                                          Jg_trapezio, g_trapezio, iteracoes)
    
    X_ = np.array(X_).T
    
    Ts.append(T)
    Xs.append(X_)
```

```python
plt.figure()

for T, X, delta in zip(Ts, Xs, [(tf - t0)/n for n in ns]):    
    plt.plot(T, X[0], label = '$\Delta t$ = '+str('%.3f'%(delta)))

plt.title('Método do Trapézio: $x_{1}(t)$ Presa-Predador')
plt.xlabel('t')
plt.ylabel('$x_{1}(t)$')

plt.legend(bbox_to_anchor=(1,1))
plt.grid(True)
plt.show()
```

Note que os gráficos ficaram bastante desiguais, por conta velocidade de convergência em função do $\Delta t$. Vamos, então, traçar os gráficos que têm formatos mais semelhantes, a partir de $\Delta t = 1.562$

```python
plt.figure()

for T, X, delta in zip(Ts[3:], Xs[3:], [(tf - t0)/n for n in ns[3:]]):    
    plt.plot(T, X[0], label = '$\Delta t$ = '+str('%.3f'%(delta)))

plt.title('Método do Trapézio: $x_{1}(t)$ Presa-Predador')
plt.xlabel('t')
plt.ylabel('$x_{1}(t)$')

plt.legend(bbox_to_anchor=(1,1))
plt.grid(True)
plt.show()
```

```python
plt.figure()
    
for T, X, delta in zip(Ts, Xs, [(tf - t0)/n for n in ns]):    
    plt.plot(T, X[1], label = '$\Delta t$ = '+str('%.3f'%(delta)))

plt.title('Método do Trapézio: $x_{2}(t)$ Presa-Predador')
plt.xlabel('t')
plt.ylabel('$x_{2}(t)$')

plt.legend(bbox_to_anchor=(1,1))
plt.grid(True)
plt.show()
```

Novamente, vemos que os gráficos ficam com escalas desiguais. Vamos, então, traçar os gráficos com escalas semelhantes a partir de $\Delta t = 1.562$

```python
plt.figure()
    
for T, X, delta in zip(Ts[3:], Xs[3:], [(tf - t0)/n for n in ns[3:]]):    
    plt.plot(T, X[1], label = '$\Delta t$ = '+str('%.3f'%(delta)))

plt.title('Método do Trapézio: $x_{2}(t)$ Presa-Predador')
plt.xlabel('t')
plt.ylabel('$x_{2}(t)$')

plt.legend(bbox_to_anchor=(1,1))
plt.grid(True)
plt.show()
```

Vemos que, em ambos os casos de $x_{1}(t)$ e $x_{2}(t)$, o Método do Trapézio converge "visualmente" por volta de $\Delta t  = 0.312$.
